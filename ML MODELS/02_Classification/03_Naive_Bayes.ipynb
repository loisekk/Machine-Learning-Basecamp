```python
# encoding all the object type columns
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
cat_cols.remove('Attrition_Flag')

df=pd.get_dummies(df,columns=cat_cols)
```

```python
df['Attrition_Flag'] = df['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1})
```

```python
# lable encoding
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['Attrition_Flag'] = le.fit_transform(df['Attrition_Flag'])
```

```python
X = df.drop("Attrition_Flag",axis=1)
y = df['Attrition_Flag']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
#
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
# object = class()
```

```python
# train
model.fit(X_train,y_train)
```

```python
# prediction
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
accuracy_score(y_train,model.predict(X_train))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d')
```

### Exploring Categorical Features vs. Attrition Flag

Let's visualize how each categorical feature is distributed with respect to the `Attrition_Flag`. This will help us understand which categories might have a higher churn rate.

```python
cat_cols_to_plot = ['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']

fig, axes = plt.subplots(len(cat_cols_to_plot), 1, figsize=(10, 5 * len(cat_cols_to_plot)))

for i, col in enumerate(cat_cols_to_plot):
    sns.countplot(data=df_original, x=col, hue='Attrition_Flag', ax=axes[i], palette='viridis')
    axes[i].set_title(f'Distribution of {col} by Attrition Flag')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### Exploring Numerical Features vs. Attrition Flag

Next, let's examine the distributions of numerical features, separated by `Attrition_Flag`. This can highlight numerical patterns associated with customer attrition.

```python
numerical_cols = df_original.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Exclude 'CLIENTNUM' and the Naive Bayes Classifier columns as they are identifiers/model outputs
numerical_cols_to_plot = [col for col in numerical_cols if col not in ['CLIENTNUM',
                                                            'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
                                                            'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']]

# Determine the number of rows needed for subplots
num_rows = (len(numerical_cols_to_plot) + 1) // 2 # +1 and //2 to round up for uneven numbers

fig, axes = plt.subplots(num_rows, 2, figsize=(15, 5 * num_rows))
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

for i, col in enumerate(numerical_cols_to_plot):
    sns.histplot(data=df_original, x=col, hue='Attrition_Flag', kde=True, ax=axes[i], palette='coolwarm')
    axes[i].set_title(f'Distribution of {col} by Attrition Flag')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

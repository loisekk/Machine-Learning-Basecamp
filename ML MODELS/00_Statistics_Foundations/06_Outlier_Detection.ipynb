```python
import numpy as np
import seaborn as sn
```

```python
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1  = np.percentile(arr , 25)
Q2  = np.percentile(arr , 50) # median
Q3  = np.percentile(arr , 75)
```

```python
IQR = Q3 - Q1 
IQR
```

## Upper Fence

```python
UF = Q3 + (1.5*IQR)
UF
```

## Lower Fence

```python
LF = Q1 - (1.5*IQR)
LF
```

```python
l = []

for i in arr:
    if i >= LF and i <= UF:
        l.append(i)
new_arr = np.array(l)
```

```python
print(arr)
print(new_arr)
```

```python
sn.boxplot(x = arr)
```

```python
sn.boxplot(new_arr)
```

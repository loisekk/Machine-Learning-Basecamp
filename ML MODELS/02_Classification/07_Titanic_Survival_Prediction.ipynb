```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sn 
import plotly.express as px 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = sn.load_dataset('titanic')
df
```

```python
df.drop(['deck', 'embark_town', 'alive' , 'who' , 'adult_male' , 'class'] , axis = 1, inplace = True )
```

```python
df.info()
```

```python
# filling none age value by mean avg 

df['age'].fillna(df['age'].mean() , inplace= True)
```

```python
df.dropna(subset=['embarked'] , inplace=True)
```

```python
df.info()
```

# label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.head()
```

```python
df['sex'] = le.fit_transform(df['sex'])
df['embarked'] = le.fit_transform(df['embarked'])
```

```python
df= df.astype(int)
df.head()
```

```python
# dividing the surviving thing to x and y label 

X = df.drop('survived' , axis = 1)
y = df['survived'] # the column we have to predict 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train , y_train)
```

```python
y_pred  = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

```python
accuracy_LR = accuracy_score(y_test , y_pred)
accuracy_LR
```

```python
import plotly.graph_objects as go

accuracy = 0.81

fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = accuracy * 100,
    title = {'text': "Model Accuracy"},
    gauge = {
        'axis': {'range': [0, 100]}
    }
))
fig.update_layout(template = 'plotly_dark')
fig.show()
```

```python
# confusion_matrix(y_test , y_pred_knn)
cm = confusion_matrix(y_test, y_pred)

fig = px.imshow(
    cm,
    text_auto=True,
    color_continuous_scale='Blues',
    title='Confusion Matrix'
)

fig.update_layout(
    xaxis_title='Predicted',
    yaxis_title='Actual',
    template = 'plotly_dark'
)

fig.show()
```

```python
print(classification_report(y_test , y_pred)) # precision , recall  , f1-score 
```

```python
# Generate classification report
report = classification_report(
    y_test,
    y_pred,
    output_dict=True
)

# Convert to DataFrame
report_df = pd.DataFrame(report).transpose()

# Keep only useful metrics
report_df = report_df.iloc[:-1, :-1]

fig = go.Figure(data=go.Heatmap(
    z=report_df.values,
    x=report_df.columns,
    y=report_df.index,
    text=report_df.round(2).values,
    texttemplate="%{text}",
    colorscale='Viridis'
))

fig.update_layout(
    title='Classification Report Heatmap',
    xaxis_title='Metrics',
    yaxis_title='Classes',
    template='plotly_dark',
    height=500
)

fig.show()
```

# Standard Scaling 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.fit_transform(X_test)
```

```python
X_train_scaler
```

```python
X_test_scaler
```

# K - Nearest Neighbors 

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors= 5)
```

```python
knn_model.fit(X_train_scaler , y_train)
```

```python
y_pred_knn = knn_model.predict(X_test_scaler)
```

```python
accuracy_score(y_test , y_pred_knn)

import plotly.graph_objects as go

accuracy = 0.80

fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = accuracy * 100,
    title = {'text': "Model Accuracy"},
    gauge = {
        'axis': {'range': [0, 100]}
    }
))
fig.update_layout(template = 'plotly_dark')
fig.show()
```

## Confusion Matrix Heatmap

```python
# confusion_matrix(y_test , y_pred_knn)

from sklearn.metrics import confusion_matrix
import plotly.express as px

cm = confusion_matrix(y_test, y_pred_knn)

fig = px.imshow(
    cm,
    text_auto=True,
    color_continuous_scale='Blues',
    title='Confusion Matrix'
)

fig.update_layout(
    xaxis_title='Predicted',
    yaxis_title='Actual',
    template = 'plotly_dark'
)

fig.show()
```

```python
print(classification_report(y_test , y_pred_knn)) # precision , recall  , f1-score 
```

##  Classification Report Heatmap

```python
import plotly.graph_objects as go
# Generate classification report
report = classification_report(
    y_test,
    y_pred_knn,
    output_dict=True
)
# Convert to DataFrame
report_df = pd.DataFrame(report).transpose()

# Keep only useful metrics
report_df = report_df.iloc[:-1, :-1]

fig = go.Figure(data=go.Heatmap(
    z=report_df.values,
    x=report_df.columns,
    y=report_df.index,
    text=report_df.round(2).values,
    texttemplate="%{text}",
    colorscale='Viridis'
))
fig.update_layout(
    title='Classification Report Heatmap',
    xaxis_title='Metrics',
    yaxis_title='Classes',
    template='plotly_dark',
    height=500
)
fig.show()
```

## K Value vs Accuracy Line Chart

```python
import plotly.graph_objects as go

k_values = list(range(1, 21))
accuracies = []

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train_scaler, y_train)

    y_pred_knn = model.predict(X_test_scaler)

    acc = accuracy_score(y_test, y_pred_knn)
    accuracies.append(acc)

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=k_values,
    y=accuracies,
    mode='lines+markers',
    name='Accuracy',
    line=dict(width=3),
    marker=dict(size=10)
))

fig.update_layout(
    title='K Value vs Accuracy',
    xaxis_title='K Value',
    yaxis_title='Accuracy',
    template='plotly_dark',
    height=500
)

fig.show()
```

## Actual vs Predicted Visualization

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    y=y_test[:100],
    mode='lines',
    name='Actual'
))

fig.add_trace(go.Scatter(
    y=y_pred_knn[:100],
    mode='lines',
    name='Predicted'
))

fig.update_layout(
    title='Actual vs Predicted',
    xaxis_title='Samples',
    yaxis_title='Class',
    template='plotly_dark'
)

fig.show()
```

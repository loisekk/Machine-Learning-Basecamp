```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sn 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sn.load_dataset('iris')
df
```

```python
print(df['species'].unique())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species',axis = 1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn =  KNeighborsClassifier(n_neighbors=13)
model_knn.fit(X_train , y_train)
```

```python
model_knn.score(X_test, y_test) # model of knn -> overfit cond
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(C= 10 , kernel= 'linear' , gamma='auto') # prev hyperparameter -> C= 30 , kernel= 'rbf'
model_svm.fit(X_train , y_train)
```

```python
model_knn.score(X_test , y_test) # -> diff parameter model Overfit
```

# Random Search CV
### Use Random Search when:
> Large Hyperparameter Space

*5 × 5 × 4 × 3 × 4
= 1200 combinations 

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_random = RandomizedSearchCV(
    
    estimator=model_svm,

    param_distributions={
        'C': [1, 10, 20, 30],  # Penalty for wrong classification
        
        'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
    },
    n_iter=4, 
    cv=5,
    
    return_train_score=False
)

classifier_random.fit(X, y)
```

```python
classifier_random = pd.DataFrame(classifier_random.cv_results_)
classifier_random.head()
```

```python
# 1- Log Transformation -> use for minimizing the Skewness  (always on numerical data) 
# Standard scaler -> mean = 0, std = 1 -> after traning -> inverse transform
# 2- SMOTE -> (synthetic minority oversampling technique) class imbalance (if there is huge difference btw two data this is called class imbalance )
```

## Log Transformation 

```python
num_col = df.select_dtypes(include=['float64','int64']).columns.tolist()
```

```python
skewness = df[num_col].skew()
```

```python
skewness
```

```python
skew_feature = skewness[abs(skewness) > 0.5].index
X_train_transformed = X_train.copy()
X_test_transformed = X_test.copy()
```

```python
for i in skew_feature:
    X_train_transformed[i] = np.log1p(X_train[i])
    X_test_transformed[i] = np.log1p(X_test[i])
```

```python
# fill the null values 
X_train_transformed.fillna(0,inplace=True)
```

## SMOTE 

```python
from imblearn.over_sampling import SMOTE
print(f"Before SMOTE:\n{y_train.value_counts()}")
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(
    X_train_transformed,
    y_train
)
print(f"After SMOTE:\n{y_train_smote.value_counts()}")
```
